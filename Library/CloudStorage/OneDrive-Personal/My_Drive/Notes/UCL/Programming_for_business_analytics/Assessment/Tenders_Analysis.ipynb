```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('EUs_Tenders_Electronic_Daily.xlsx', engine='openpyxl')

# To check your data:
print(df.head())  # Shows first 5 rows
print(df.shape)   # Shows number of rows and columns
```

```python
print(df.columns)
```

```python
columns_to_keep = ['persistent_id', 'tender_title', 'tender_proceduretype', 'tender_supplytype', 
                  'tender_isjointprocurement', 'tender_cpvs', 'tender_iseufunded', 
                  'tender_selectionmethod', 'tender_finalprice', 'buyer_masterid', 'buyer_name',
                  'buyer_nuts', 'buyer_mainactivities', 'buyer_buyertype', 'bidder_masterid',
                  'bidder_name', 'bidder_country_e', 'bid_issubcontracted', 
                  'bid_subcontractedproportion', 'bid_isconsortium', 'tender_year',
                  'tender_description_length', 'tender_personalrequirements_length',
                  'tender_technicalrequirements_length', 'tender_economicrequirements_length',
                  'bid_priceUsd', 'tender_estimatedpriceUsd', 'tender_finalpriceUsd',
                  'corr_singleb', 'corr_proc', 'submission_period', 'corr_subm', 'corr_nocft',
                  'decision_period', 'corr_decp', 'corr_buyer_concentration', 'cri']

df = df[columns_to_keep]

df.head(5)
```

```python
print(df.isnull().sum())
```

```python
print(df.dtypes)
```

```python
# For object (string/text) columns
df['persistent_id'] = df['persistent_id'].fillna('Unknown')
df['tender_title'] = df['tender_title'].fillna('Unknown')
df['tender_proceduretype'] = df['tender_proceduretype'].fillna('Not Specified')
df['tender_supplytype'] = df['tender_supplytype'].fillna('Not Specified')
df['tender_isjointprocurement'] = df['tender_isjointprocurement'].fillna('Unknown')
df['tender_cpvs'] = df['tender_cpvs'].fillna('Not Specified')
df['tender_iseufunded'] = df['tender_iseufunded'].fillna('Unknown')
df['tender_selectionmethod'] = df['tender_selectionmethod'].fillna('Not Specified')
df['buyer_masterid'] = df['buyer_masterid'].fillna('Unknown')
df['buyer_name'] = df['buyer_name'].fillna('Unknown')
df['buyer_nuts'] = df['buyer_nuts'].fillna('Not Specified')
df['buyer_mainactivities'] = df['buyer_mainactivities'].fillna('Unknown')
df['buyer_buyertype'] = df['buyer_buyertype'].fillna('Unknown')
df['bidder_masterid'] = df['bidder_masterid'].fillna('Unknown')
df['bidder_name'] = df['bidder_name'].fillna('Unknown')
df['bidder_country_e'] = df['bidder_country_e'].fillna('Unknown')
df['bid_issubcontracted'] = df['bid_issubcontracted'].fillna('Unknown')
df['bid_isconsortium'] = df['bid_isconsortium'].fillna('Unknown')

# For float64 columns (numerical with decimals)
df['tender_finalprice'] = df['tender_finalprice'].fillna(df['tender_finalprice'].mean())
df['bid_subcontractedproportion'] = df['bid_subcontractedproportion'].fillna(0)
df['bid_priceUsd'] = df['bid_priceUsd'].fillna(df['bid_priceUsd'].mean())
df['tender_estimatedpriceUsd'] = df['tender_estimatedpriceUsd'].fillna(df['tender_estimatedpriceUsd'].mean())
df['tender_finalpriceUsd'] = df['tender_finalpriceUsd'].fillna(df['tender_finalpriceUsd'].mean())
df['corr_singleb'] = df['corr_singleb'].fillna(df['corr_singleb'].mean())
df['corr_proc'] = df['corr_proc'].fillna(df['corr_proc'].mean())
df['submission_period'] = df['submission_period'].fillna(df['submission_period'].mean())
df['decision_period'] = df['decision_period'].fillna(df['decision_period'].mean())
df['corr_decp'] = df['corr_decp'].fillna(df['corr_decp'].mean())
df['corr_buyer_concentration'] = df['corr_buyer_concentration'].fillna(df['corr_buyer_concentration'].mean())
df['cri'] = df['cri'].fillna(df['cri'].mean())

# int64 columns (integers) don't need handling as they show 0 nulls:
# tender_year, tender_description_length, tender_personalrequirements_length,
# tender_technicalrequirements_length, tender_economicrequirements_length,
# corr_subm, corr_nocft

# Check if any nulls remain
print(df.isnull().sum())
```
